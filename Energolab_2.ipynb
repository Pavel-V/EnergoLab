### Подключение библиотек

```
import numpy as np
import pandas as pd
import requests
import io
import matplotlib.pyplot as plt
```

### Чтение данных о частоте

```
url = "https://drive.google.com/uc?export=download&id=1tV4oEN14K3UkQP4fq6y9no7JGjCm_iVu"
content = requests.get(url).content
input_df_source = pd.read_csv(io.StringIO(content.decode('utf-8')))
```

```
input_df_source.head()
```

```
input_df_source.columns
```

### Построение графика

```
fig, ax = plt.subplots(figsize=(25, 10)) 

ax.plot(np.arange(len(input_df_source)), input_df_source['coal'], 'black')
ax.plot(np.arange(len(input_df_source)), input_df_source['wind'], 'green')
ax.plot(np.arange(len(input_df_source)), input_df_source['hydro'], 'blue')
ax.plot(np.arange(len(input_df_source)), input_df_source['nuclear'], 'red')

plt.legend(['coal', 'wind', 'hydro', 'nuclear'])

ax.set_xlabel('Time') 
ax.set_ylabel('Amplitude')

plt.show()
```

### Пример круговой диаграммы

```
labels = 'A', 'B', 'C', 'D'
values = [15, 30, 45, 10]
explode = (0.1, 0.1, 0.1, 0.1)

fig, ax = plt.subplots()
ax.pie(values, explode = explode, labels = labels, autopct = '%1.1f%%', shadow = True, startangle = 90)
ax.axis('equal') 

plt.show()
```
