### Подключение библиотек

```
import numpy as np
import pandas as pd
import requests
import io
import matplotlib.pyplot as plt
```

### Чтение данных о частоте

```
url = "https://drive.google.com/uc?export=download&id=1-JZfsCiKbxXOHDSjkiEr2E0u69DwVhq0"
content = requests.get(url).content
input_df_freq = pd.read_csv(io.StringIO(content.decode('utf-8')), header = None)
```

```
input_df_freq.head()
```

Извлечение столбца со значениями частоты

```
frequency = input_df_freq[2]
frequency.describe()
```

### Очищение данных - удаление нулевых значений

```
frequency_cleared = frequency[frequency > 0.001]
frequency_cleared.describe()
```

```
freq_arr = frequency_cleared.values
print(freq_arr)
```

### Построение графика

```
fig, ax = plt.subplots(figsize = (30, 10)) 

ax.plot(np.arange(len(freq_arr)), freq_arr, 'red')
ax.set_xlabel('Time') 
ax.set_ylabel('Amplitude')

plt.show()
```
